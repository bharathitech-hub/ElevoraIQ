### 4.1. Clean Column Names

To streamline preprocessing and coding readability, all column names are standardized by replacing spaces with underscores.


```python
# Standardize column names
df.columns = df.columns.str.replace(' ', '_')
```

### 5. Encode Ordered Categorical Features

We use Label Encoding for features that have an inherent order. In this case:
- `Soft_Skills`: Ordered as Average < Good < Excellent
- `Notice_Period`: Ordered by urgency (Immediate < 30 Days < 60 Days)

Label encoding assigns a numerical value to each category based on order.


```python
from sklearn.preprocessing import LabelEncoder

# Initialize label encoder
le = LabelEncoder()

# Encode ordered features
df['Soft_Skills'] = le.fit_transform(df['Soft_Skills'])
df['Notice_Period'] = le.fit_transform(df['Notice_Period'])
```

### 5.1. Encode Unordered Categorical Features

For features with no natural order, such as `Education_Level`, `Gender`, or `Job_Role`, we apply One-Hot Encoding.  
We also use `drop_first=True` to avoid the dummy variable trap.


```python
# List of unordered categorical features
categorical_columns = [
    'Education_Level', 'Certifications', 'Industry', 'Location',
    'Job_Role', 'Company_Type', 'Gender', 'Work_Gap_Status'
]

# One-hot encode these features
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

### 6. Feature Engineering: Skills Column

The `Skills` column contains multiple skills in a single string (separated by `;`). We:
- Split and count each unique skill
- Select only those skills that occur more than 5 times
- Create binary columns like `has_Python`, `has_SQL` to indicate presence


### 6.1 Feature Engineering

To enhance model performance and ensure compatibility with machine learning algorithms, the following feature engineering strategies were applied:

- **Categorical Variable Encoding**  
  All categorical features (e.g., job role, industry domain, education level) were transformed using One-Hot Encoding. This method preserves the nominal nature of the data without introducing unintended ordinal relationships.

- **Numerical Feature Scaling**  
  Continuous variables were standardized using `StandardScaler` to normalize feature distributions. This was particularly important for algorithms sensitive to scale, such as Support Vector Regression.

These transformations ensured a consistent data structure suitable for both linear and non-linear model architectures.


```python
# Count individual skill frequency
skill_counts = df['Skills'].str.split('; ').explode().value_counts()

# Select skills used by at least 5 candidates
common_skills = skill_counts[skill_counts > 5].index

# Create binary columns for common skills
for skill in common_skills:
    skill_col = 'has_' + skill.replace(' ', '_')
    df[skill_col] = df['Skills'].apply(lambda x: 1 if skill in x else 0)

# Drop the original Skills column
df.drop('Skills', axis=1, inplace=True)
```

### 7. Feature Scaling

Feature scaling is crucial to ensure that all numeric features contribute equally to the model.  
This step becomes especially important when using distance-based algorithms (e.g., KNN, SVM).

We use **StandardScaler** to standardize features by removing the mean and scaling to unit variance.


```python
from sklearn.preprocessing import StandardScaler

# Initialize scaler
scaler = StandardScaler()

# Select only numerical columns for scaling
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply scaling
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

### 7.1. Train-Test Split

Before model training, we split the dataset into training and testing subsets (80/20 split).  
This allows us to evaluate model performance on unseen data.


```python
from sklearn.model_selection import train_test_split

# Define features and target
x = df.drop("Expected_Salary", axis=1) # Changed 'Salary' to 'Expected_Salary'
y = df["Expected_Salary"] # Changed 'Salary' to 'Expected_Salary'

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### 8. Train Multiple Regression Models

We train and evaluate four regression models:
- Linear Regression
- Random Forest Regressor
- XGBoost Regressor
- Support Vector Regressor (SVR)

Each model will be evaluated using R², MAE, and RMSE.


### 8.1 Model Selection Justification

To ensure a balanced evaluation of both linear and non-linear relationships in the data, the following models were selected:

- **Linear Regression**  
  Used as a benchmark model to establish a baseline with interpretable linear trends.

- **Random Forest Regressor**  
  Chosen for its ability to capture complex, non-linear interactions and provide feature importance insights.

- **XGBoost Regressor**  
  Selected for its high performance on structured data and built-in handling of overfitting through regularization.

- **Support Vector Regressor (SVR)**  
  Applied to assess kernel-based learning effectiveness on compensation prediction with limited data size.


```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
```

```python
# Initialize models
lr = LinearRegression()
rf = RandomForestRegressor(random_state=42)
```

```python
# Fit the models to the training data
lr.fit(x_train, y_train) # Fit Linear Regression
rf.fit(x_train, y_train) # Fit Random Forest
```

```python
# Predict
lr_preds = lr.predict(x_test)
rf_preds = rf.predict(x_test)
```

```python
# Evaluation function
def evaluate_model(name, y_true, y_pred):
    print(f"{name} Evaluation:")
    print(f"R² Score: {r2_score(y_true, y_pred):.2f}")
    print(f"MAE: {mean_absolute_error(y_true, y_pred):.2f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_true, y_pred)):.2f}")
    print("-" * 30)
```

```python
# Evaluate both
evaluate_model("Linear Regression", y_test, lr_preds)
evaluate_model("Random Forest", y_test, rf_preds)
```

```python
from xgboost import XGBRegressor

# Initialize and train
xgb = XGBRegressor(random_state=42)
xgb.fit(x_train, y_train)

# Predict
xgb_preds = xgb.predict(x_test)

# Evaluate
evaluate_model("XGBoost Regressor", y_test, xgb_preds)
```

```python
from sklearn.svm import SVR

# Initialize and train
svr = SVR()
svr.fit(x_train, y_train)

# Predict
svr_preds = svr.predict(x_test)

# Evaluate
evaluate_model("Support Vector Regressor", y_test, svr_preds)
```

##8.2 Why Classification Metrics Are Not Used

This project involves predicting the expected salary of candidates based on a combination of numerical and categorical features. Since the target variable is continuous, the problem falls under the category of regression rather than classification.

Metrics such as precision, recall, and F1-score are specifically designed for classification problems where the objective is to predict discrete class labels. These metrics are effective in evaluating how well a model distinguishes between different categories or outcomes.

In contrast, regression tasks aim to predict continuous values. Therefore, the performance of regression models is evaluated using appropriate metrics such as:

- **R² Score (Coefficient of Determination):** Indicates the proportion of variance in the dependent variable that is predictable from the independent variables.
- **Root Mean Squared Error (RMSE):** Measures the standard deviation of prediction errors, providing insight into the average magnitude of errors.
- **Mean Absolute Error (MAE):** Represents the average absolute difference between predicted values and actual values.

Applying classification metrics to a regression problem would be conceptually incorrect and may lead to misleading interpretations. Hence, only regression-specific evaluation metrics are utilized in this project to ensure accurate and meaningful performance assessment.


```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Models and their R² scores
models = ["Linear Regression", "Random Forest", "XGBoost", "SVR"]
r2_scores = [
    r2_score(y_test, lr_preds),
    r2_score(y_test, rf_preds),
    r2_score(y_test, xgb_preds),
    r2_score(y_test, svr_preds)
]

# Bar chart
plt.bar(models, r2_scores, color='skyblue')
plt.title("Model Comparison - R² Score")
plt.ylabel("R² Score")
plt.ylim(0, 1)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

# Models and MAE values
mae_scores = [
    mean_absolute_error(y_test, lr_preds),
    mean_absolute_error(y_test, rf_preds),
    mean_absolute_error(y_test, xgb_preds),
    mean_absolute_error(y_test, svr_preds)
]

# Bar chart
plt.bar(models, mae_scores, color='lightcoral')
plt.title("Model Comparison - MAE")
plt.ylabel("Mean Absolute Error")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Models and RMSE values
rmse_scores = [
    np.sqrt(mean_squared_error(y_test, lr_preds)),
    np.sqrt(mean_squared_error(y_test, rf_preds)),
    np.sqrt(mean_squared_error(y_test, xgb_preds)),
    np.sqrt(mean_squared_error(y_test, svr_preds))
]

# Bar chart
plt.bar(models, rmse_scores, color='mediumseagreen')
plt.title("Model Comparison - RMSE")
plt.ylabel("Root Mean Squared Error")
plt.show()
```
